```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Seed for reproducibility
np.random.seed(42)

# Parameters
n = 40
p = 0.5
num_samples = 100

# Step 1: Generate Samples
samples = np.random.binomial(n, p, num_samples)

# Step 2: Estimate Mean and Variance
sample_mean = np.mean(samples)
sample_variance = np.var(samples)

# Step 3: Visualization
# Theoretical PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plotting
plt.figure(figsize=(12, 6))

# Histogram of the sample data
plt.hist(samples, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='g', label='Sample Data')

# PMF of the binomial distribution
plt.plot(x, pmf, 'bo', ms=8, label='Binomial PMF')
plt.vlines(x, 0, pmf, colors='b', lw=5, alpha=0.5)

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})\nSample Mean: {sample_mean:.2f}, Sample Variance: {sample_variance:.2f}')
plt.legend()
plt.grid(True)
plt.show()
```
