## Introduction to Monte Carlo Simulation for Estimating Area Inside Rose Petals

Consider the rose curve defined by the equation:

$$ r = a \cos(k\theta) $$

where $ r $ is the radial distance from the origin, $ \theta $ is the angle parameter ranging from $ 0 $ to $ 2\pi $, $ a $ determines the size of the rose, and $ k $ determines the number of petals.

For more details, you can refer to the [Wikipedia page on Rose (mathematics)](https://en.wikipedia.org/wiki/Rose_(mathematics)).


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1  # Amplitude of the rose curve
k = 2  # Number of petals, k = 2 for 4 petals

# Generate values for theta from 0 to 2*pi
theta = np.linspace(0, 2*np.pi, 1000)

# Calculate r for each theta
r = a * np.cos(k * theta)

# Convert polar coordinates (r, theta) to Cartesian coordinates (x, y)
x = r * np.cos(theta)
y = r * np.sin(theta)

# Plotting the rose curve
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='blue', linewidth=2, label=f'Rose curve with {4} petals')
plt.title(f'Rose Curve with {4} Petals')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')  # Ensure x and y axes have the same scale
plt.legend()
plt.show()
```
