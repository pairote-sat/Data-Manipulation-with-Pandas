````python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Seed for reproducibility
np.random.seed(42)

# Parameters
n = 3
p = 0.89
num_samples = 100

# Generate samples
samples = np.random.binomial(n, p, num_samples)
print("Generated Samples:", samples)

# Calculate mean and variance
mean_sample = np.mean(samples)
variance_sample = np.var(samples)
print(f"Sample Mean: {mean_sample}")
print(f"Sample Variance: {variance_sample}")

# Theoretical PMF
x = np.arange(0, n+1)
theoretical_pmf = binom.pmf(x, n, p)

# Sample PMF
sample_pmf, bins = np.histogram(samples, bins=np.arange(0, n+2), density=True)

# Plot
plt.bar(x - 0.2, sample_pmf, width=0.4, label='Sample PMF')
plt.bar(x + 0.2, theoretical_pmf, width=0.4, label='Theoretical PMF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Comparison of Sample PMF and Theoretical PMF')
plt.legend()
plt.show()

# Estimate probability P(X = 3)
prob_X_eq_3 = np.sum(samples == 3) / num_samples
print(f"Estimated P(X = 3): {prob_X_eq_3}")


import matplotlib.pyplot as plt
from scipy.stats import binom

# Seed for reproducibility
np.random.seed(42)

# Parameters
n = 3
p = 0.89
num_samples = 100

# Generate samples
samples = np.random.binomial(n, p, num_samples)
print("Generated Samples:", samples)

# Calculate mean and variance
mean_sample = np.mean(samples)
variance_sample = np.var(samples)
print(f"Sample Mean: {mean_sample}")
print(f"Sample Variance: {variance_sample}")

# Theoretical PMF
x = np.arange(0, n+1)
theoretical_pmf = binom.pmf(x, n, p)

# Sample PMF
sample_pmf, bins = np.histogram(samples, bins=np.arange(0, n+2), density=True)

# Plot
plt.bar(x - 0.2, sample_pmf, width=0.4, label='Sample PMF')
plt.bar(x + 0.2, theoretical_pmf, width=0.4, label='Theoretical PMF')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Comparison of Sample PMF and Theoretical PMF')
plt.legend()
plt.show()

# Estimate probability P(X = 3)
prob_X_eq_3 = np.sum(samples == 3) / num_samples
print(f"Estimated P(X = 3): {prob_X_eq_3}")
````

```{admonition} Question 2: Monte Carlo Simulation Exercise: Poisson Distribution

**Background:**

Customers arrive at a photocopying machine at an average rate of 2 every five minutes. Assume that these arrivals are independent, with a constant arrival rate, and that this problem follows a Poisson model, with \( X \) denoting the number of arriving customers in a 5-minute period and mean \( \lambda = 2 \).

**Objective:**

Use Monte Carlo simulation to estimate the probability that more than two customers arrive in a 5-minute period.

How can you generate 1000 samples from a Poisson distribution with \( \lambda = 2 \) using the Monte Carlo method? Provide the Python code used to generate these samples.

How can you visualize the comparison between the sample data and the theoretical probability mass function (PMF) of the Poisson distribution? Provide the Python code used for the visualization and describe the resulting plot.

4. **Estimate Probability**

Using the Monte Carlo results, estimate the probability that more than two customers arrive in a 5-minute period. Provide the Python code and the estimated probability.

Here is the Python code to solve Question 2.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Seed for reproducibility
np.random.seed(42)

# Parameters
lambda_ = 2
num_samples = 1000

# Question 1: Generate Samples
samples = np.random.poisson(lambda_, num_samples)
print("Generated Samples:", samples)

# Question 2: Estimate Mean and Variance
mean_sample = np.mean(samples)
variance_sample = np.var(samples)
print(f"Sample Mean: {mean_sample}")
print(f"Sample Variance: {variance_sample}")

# Question 3: Visualization
# Theoretical PMF
x = np.arange(0, max(samples)+1)
theoretical_pmf = poisson.pmf(x, lambda_)

# Sample PMF
sample_pmf, bins = np.histogram(samples, bins=np.arange(0, max(samples)+2), density=True)

# Plot
plt.bar(x - 0.2, sample_pmf, width=0.4, label='Sample PMF')
plt.bar(x + 0.2, theoretical_pmf, width=0.4, label='Theoretical PMF')
plt.xlabel('Number of Arrivals')
plt.ylabel('Probability')
plt.title('Comparison of Sample PMF and Theoretical PMF')
plt.legend()
plt.show()

# Question 4: Estimate Probability
prob_more_than_two = np.sum(samples > 2) / num_samples
print(f"Estimated Probability (X > 2): {prob_more_than_two}")


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Seed for reproducibility
np.random.seed(42)

# Parameters
lambda_ = 2
num_samples = 1000

# Question 1: Generate Samples
samples = np.random.poisson(lambda_, num_samples)
print("Generated Samples:", samples)

# Question 2: Estimate Mean and Variance
mean_sample = np.mean(samples)
variance_sample = np.var(samples)
print(f"Sample Mean: {mean_sample}")
print(f"Sample Variance: {variance_sample}")

# Question 3: Visualization
# Theoretical PMF
x = np.arange(0, max(samples)+1)
theoretical_pmf = poisson.pmf(x, lambda_)

# Sample PMF
sample_pmf, bins = np.histogram(samples, bins=np.arange(0, max(samples)+2), density=True)

# Plot
plt.bar(x - 0.2, sample_pmf, width=0.4, label='Sample PMF')
plt.bar(x + 0.2, theoretical_pmf, width=0.4, label='Theoretical PMF')
plt.xlabel('Number of Arrivals')
plt.ylabel('Probability')
plt.title('Comparison of Sample PMF and Theoretical PMF')
plt.legend()
plt.show()

# Question 4: Estimate Probability
prob_more_than_two = np.sum(samples > 2) / num_samples
print(f"Estimated Probability (X > 2): {prob_more_than_two}")
````
