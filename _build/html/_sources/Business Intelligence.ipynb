## Descriptive Analytics: Data Visualization
Descriptive analytics can be effectively performed using Python with libraries designed for data visualization.

### Tools:
- **Matplotlib**: A comprehensive library for creating static, animated, and interactive visualizations.
- **Seaborn**: A Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.
- **Plotly**: An interactive graphing library that makes interactive, publication-quality graphs online.

### Example:
```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sample data
data = sns.load_dataset("iris")

# Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(data['sepal_length'], bins=20, color='blue')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Seaborn
sns.pairplot(data, hue='species')
plt.show()

# Plotly
fig = px.scatter(data, x='sepal_length', y='sepal_width', color='species')
fig.show()


````python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sample data
data = sns.load_dataset("iris")

# Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(data['sepal_length'], bins=20, color='blue')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Seaborn
sns.pairplot(data, hue='species')
plt.show()

# Plotly
fig = px.scatter(data, x='sepal_length', y='sepal_width', color='species')
fig.show()
````

# Diagnostic Analytics: Monte Carlo Simulation

Monte Carlo Simulation can be conducted in Python using libraries that support statistical computations and simulations.

## Tools:
- **NumPy**: A library for numerical computing with powerful capabilities for random number generation.
- **SciPy**: A library used for scientific and technical computing.
- **Pandas**: A data manipulation library that provides data structures and functions needed to manipulate numerical tables and time series.

## Example:
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Monte Carlo Simulation for predicting future stock prices
def monte_carlo_simulation(start_price, days, mu, sigma, simulations):
    results = []
    for _ in range(simulations):
        prices = [start_price]
        for _ in range(days):
            prices.append(prices[-1] * np.exp(np.random.normal(mu, sigma)))
        results.append(prices)
    return np.array(results)

# Parameters
start_price = 100
days = 365
mu = 0.0002
sigma = 0.01
simulations = 1000

# Run simulation
simulated_prices = monte_carlo_simulation(start_price, days, mu, sigma, simulations)

# Convert results to DataFrame for analysis
simulated_prices_df = pd.DataFrame(simulated_prices.T)

# Plot results
simulated_prices_df.plot(legend=False)
plt.title('Monte Carlo Simulation of Stock Prices')
plt.xlabel('Days')
plt.ylabel('Price')
plt.show()


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Monte Carlo Simulation for predicting future stock prices
def monte_carlo_simulation(start_price, days, mu, sigma, simulations):
    results = []
    for _ in range(simulations):
        prices = [start_price]
        for _ in range(days):
            prices.append(prices[-1] * np.exp(np.random.normal(mu, sigma)))
        results.append(prices)
    return np.array(results)

# Parameters
start_price = 100
days = 365
mu = 0.0002
sigma = 0.01
simulations = 1000

# Run simulation
simulated_prices = monte_carlo_simulation(start_price, days, mu, sigma, simulations)

# Convert results to DataFrame for analysis
simulated_prices_df = pd.DataFrame(simulated_prices.T)

# Plot results
simulated_prices_df.plot(legend=False)
plt.title('Monte Carlo Simulation of Stock Prices')
plt.xlabel('Days')
plt.ylabel('Price')
plt.show()
````
